```python
import numpy as np
import pandas as pd
import os
from glob import glob
import matplotlib.pyplot as plt
from tqdm import tqdm

# レースデータの読み込み

race_data_dir = "../../jra_race_data"
# 2018年以上のデータを取得
# （2000_01_05_水曜_10_25_1回中山1日_サラ系4歳未勝利.csv）形式で保存されている
# （年_月_日_曜日_時刻_回数_場所_日数_レース名.csv）
year_2018_race_dir_template = os.path.join(race_data_dir, "2018", "*", "*", "*.csv")
year_2019_race_dir_template = os.path.join(race_data_dir, "2019", "*", "*", "*.csv")
year_over_2020_race_dir_template = os.path.join(race_data_dir, "202*", "*", "*", "*.csv")
race_data_path_list = glob(year_2018_race_dir_template) + glob(year_2019_race_dir_template) + glob(year_over_2020_race_dir_template)

print(len(race_data_path_list))

# レースデータフレームの作成
for n_race, race_data_path in tqdm(enumerate(race_data_path_list)):
    if n_race == 0:
        new_race_df = pd.read_csv(race_data_path)
    else:
        tmp_new_race_df = pd.read_csv(race_data_path)
        new_race_df = pd.concat([new_race_df, tmp_new_race_df], axis=0)

    # 確認用のため、1000レースにつき1回出力
    if n_race % 5000 == 0:
        print(race_data_path)    

```

```python
# 馬場データの読み込み

baba_data_dir = "jra_baba_data"
csv_dir_template = os.path.join(baba_data_dir, "csv", "*", "*.csv")

baba_data_path_list = glob(csv_dir_template)

for n_baba, baba_data_path in tqdm(enumerate(baba_data_path_list)):
    if n_baba == 0:
        new_baba_df = pd.read_csv(baba_data_path)
    else:
        tmp_new_baba_df = pd.read_csv(baba_data_path)
        new_baba_df = pd.concat([new_baba_df, tmp_new_baba_df], axis=0)

    # 確認用のため、1000レースにつき1回出力
    if n_baba % 10 == 0:
        print(baba_data_path)
```

```python
# レースデータの列名を確認

race_df = new_race_df
race_df_columns = race_df.columns

# 馬場データの列名を確認

baba_df = new_baba_df
baba_df_columns = baba_df.columns

# 中身を確認
print(race_df_columns)
print()
print(baba_df_columns)
```

```python
# 利益を計算
# prediction_labelが0の場合、単勝オッズを掛けた金額を配当金として加算
# prediction_labelが1の場合は無視

# 馬券の購入金額は100円とする


buy_price = 10000

# 利益
total_dividend = 0

# 買った回数と負けた回数のカウント
buy_count = 0
lose_count = 0

# 買った際のオッズのリスト
odds_list = []

# 利益のリスト
dividend_list = []
dividend_list.append(0)

# 比較用の前回合計利益
prev_dividend = 0

for index, row in result.iterrows():
    if row["prediction_label"] == 0:
        print("*" * 100)
        if row["target"] == 0:
            total_dividend += row["単勝オッズ"] * buy_price - buy_price
            buy_count += 1
            win_odds = row["単勝オッズ"]
            odds_list.append(win_odds)
            print(f"利益発生！ {prev_dividend:.0f}円 -> {total_dividend:.0f}円")
        else:
            total_dividend -= buy_price
            lose_count += 1
            print(f"損失発生！ {prev_dividend:.0f}円 -> {total_dividend:.0f}円")
        dividend_list.append(total_dividend)
    else:
        pass
    
    # 小数点第1位を四捨五入

    prev_dividend = total_dividend

```

```python
# オッズが1未満の数をカウント
under_one_count = 0
for odds in odds_list:
    if odds < 1:
        under_one_count += 1

print("レース数: ", len(dividend_list) - 1)
print("購入回数: ", buy_count)
print("負けた回数: ", lose_count)
print("オッズが1未満の数: ", under_one_count)
print("利益: ", total_dividend)

# 利益のリストを可視化
plt.plot(dividend_list)
plt.xlabel("レース回数")
plt.ylabel("利益")
plt.show()

# オッズのリストをヒストグラムで可視化
plt.hist(odds_list, bins=1000)
plt.xlabel("オッズ")
plt.ylabel("回数")
plt.show()

# ヒストグラムの範囲が広すぎるため、オッズが50以下のデータのみを可視化
# 見やすいようにgridを追加
# x軸に数値を追加
plt.hist(odds_list, bins=50, range=(0, 50))
plt.xlabel("オッズ（50以下）")
plt.ylabel("回数")

# x軸に数値を追加
plt.xticks(np.arange(0, 51, 1))
plt.show()
```
